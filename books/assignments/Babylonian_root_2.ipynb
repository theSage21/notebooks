# Babylonian method

[Wikipedia](https://en.wikipedia.org/wiki/Methods_of_computing_square_roots#Babylonian_method)

Here we show:
 - how error falls off with iterations
 - how the value approaches the true value with iterations

```python
values, errors = [], []
def babylon(n, max_iters=100, tolerance=1e-10):
    values.append(n)
    x = ( n + (2 / n))/2
    err = n - x
    global errors
    errors.append(err)
    if (max_iters < 1) or (abs(err) < tolerance):
        return x
    else:
        return babylon(x, max_iters - 1, tolerance)

root2 = babylon(0.5)
print('Calculated: ', root2)
print('Actual:     ', math.sqrt(2))
print('Difference: ', math.sqrt(2) - root2)
```

```python
plt.plot([abs(i) for i in errors])
plt.xlabel('Iteration')
plt.ylabel('Absolute Error')
plt.title('Error falloff with iterations')
```

We can see that the error falls off quickly with the iterations. Here we have achieved `1e-10` accuracy for the value of `√2`

```python
figsize=(15, 6)
plt.figure(figsize=figsize)
plt.subplots(12, figsize=figsize)
for guess in range(1, 20):
    guess = guess / 10.0
    values, errors = [], []
    label = 'Guess = {}'.format(guess)
    pi = babylon(guess)
    plt.subplot(121)
    plt.plot([abs(i) for i in errors], label=label)
    plt.subplot(122)
    plt.plot(values, label=label)
plt.subplot(121)
plt.title('Convergence of absolute as guesses change')
plt.xlabel('Iteration')
plt.ylabel('Absolute Error')
plt.subplot(122)
plt.title('Convergence of Estimated value with iterations')
plt.xlabel('Iteration')
plt.ylabel('Estimated value of √2')
plt.legend()
```

Now we also see that no matter if the initial estimate is higher or lower than the true value, the function converges to the needed value with enough iterations.
