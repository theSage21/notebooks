```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%pylab inline
```

# Vander Pol Oscillation
The equation under consideration is $$\frac{d^2x}{dt^2} - \mu(1-x^2)\frac{dx}{dt} + x = 0$$

Since they are not of first order, we do the following.

$\mu$ is a constant. If we let $y=x - x^3/3$ we arrive at the following equations:
$$\frac{dx}{dt} = \mu(x-1/3x^3-y)$$

$$\frac{dy}{dt} = \mu/x$$

```python
mu = 1.0

def vanderpol(X, t):
    x = X[0]
    y = X[1]
    dxdt = mu * (x - 1./3.*x**3 - y)
    dydt = x / mu
    return [dxdt, dydt]

X0 = [1, 2]
t = np.linspace(0, 40, 250)

sol = odeint(vanderpol, X0, t)

x = sol[:, 0]
y = sol[:, 1]

plt.figure(figsize=(10, 7))
plt.plot(t,x, t, y)
plt.xlabel('t')
plt.legend(('x', 'y'))
```

```python
# phase portrait
plt.figure(figsize=(10, 10))
plt.plot(x,y)
plt.plot(x[0], y[0], 'ro')
plt.xlabel('x')
plt.ylabel('y')
```
