```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
%pylab inline
```

# Simple pendulum

The equation $$\theta = A\sin{\sqrt{\frac{g}{L}}t}$$ describes the system. 

```python
g = 9.8
L = 1.0 # physical length of bar
A = 1.0 #amplitude
```

```python
t = np.linspace(0., 10, 1000)
theta = A * np.sin(np.sqrt(g/L) * t)
```

```python
plt.plot(t, theta)
plt.xlabel('Time')
plt.ylabel('Theta of the pendulum angle.')
plt.title('Simple Pendulum')
```
