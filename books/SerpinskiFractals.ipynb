```python
from tqdm import tqdm, trange
import pandas as pd
%pylab inline
```

```python
r = random.random
scale = 10
def midway(a, b):
    return ((a[0]+b[0])/2, (a[1]+b[1])/2)

end_points = [(r()*scale, r()*scale),
              (r()*scale, r()*scale),
              (r()*scale, r()*scale)
             ]
```

```python
points = [(r()*scale, r()*scale)]
for _ in trange(50000):
    last = points[-1]
    choice = random.choice(range(len(end_points)))
    choice = end_points[choice]
    new_point = midway(last, choice)
    points.append(new_point)
```

```python
df = pd.DataFrame(points, columns=['X','Y'])
plt.subplots(figsize=(15, 15))
for index, limit in enumerate([100, 500, 1000, 10000, 20000, 50000]):
    plt.subplot(3, 3, index+1)
    plt.scatter(df[:limit].X, df[:limit].Y, marker='.')
    plt.axis('off')
    plt.title('{} iterations'.format(limit))
```
