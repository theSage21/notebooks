```python
x = np.arange(-1, 2, 0.0001)
x = np.array([np.complex(i) for i in x])
a = np.power(x, x)

y = [float(i.real) for i in a]
z = [float(i.imag) for i in a]

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, label='x^x curve')
ones = np.array(x)
zs = np.zeros(x.shape[0])
ax.plot(ones, zs, zs, label='X axis')
ax.plot(zs, ones, zs, label='Y-real')
ax.plot(zs, zs, ones, label='Y-imaginary')
plt.legend()
```
