```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
%matplotlib notebook
```

```python
def poincare_dist(a, b):
    num = 2 * np.square(a-b)
    deno = (1 - np.square(a)) * (1 - np.square(b))
    dist = np.arccosh( 1 + (num / deno))
    return dist

def linear_dist(a, b):
    return np.linalg.norm(a-b, 2)
```

```python
points = np.random.random((10000, 2))*-2 + 1
inside_circle = np.linalg.norm(points, 2, axis=1) < 1
p = points[inside_circle]
p.shape
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(p[:, 0], p[:, 1], marker='.')
```

```python
hyp = np.array([poincare_dist(i, np.zeros(2)) for i in p])
hyp.shape
```

```python
plt.scatter(p[:,0], hyp[:,0], marker='.')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(p[:,0], p[:,1], hyp[:,1])
```
